# 任务内容

## 需要了解的包
1. Pandas  用于数据分析和处理, 可以支持多维cube
2. matplotlib  python中用于绘图的包, 比较简单(缺点是api比较混乱, 写的比较差)
3. Seaborn 可以画很多高级的图像, 比如热量图等等. (可以现查)
4. sklearn 传统机器学习包, 有大量机器学习库
5. XGBoost 经典集成学习库
6. pytorch 主流深度学习库

## 需要做的分析

### 数据预处理
1. 对原始数据特征的描述: 如均值, 方差, 偏度, 取值范围, 图像, 缺失值等等(部分可以用data.summary实现)
2. 离群值的检测: 箱型图等
3. 数据可视化(target和features的pair plot, distribution plot)
4. 数据之间相关性的可视化: 一方面是相关系数图, 另一方面是条件分布图(可以用风琴图)
6. 其他觉得可行的预处理和可视化

### 特征提取
1. k-means检测聚类, 可视化聚类特征
2. 经典PCA方法
3. 其他可行的手段(例如将相关系数较高的features进行合成)

### 模型构建
1. logistic 回归
2. SVM
3. 树和森林家族
4. CNN/LSTM

### 连接github
相信怎么连接大家都能在网上找到, 我就不再多说了. 这里要提醒的是
1. 每个人都建一个自己的分支, 不要在主分支上直接修改!!!
2. 每次在自己的分支上提交完以后, 记得到github网页端合并更改. 确保无冲突后可merge
(当然也可以在本地合并更改后在main分支提交(这种方法不需要去网页端合并), 但不建议这么做)
3. 每次更改完记得pull回本地

### 使用Jupyter notebook
1. Jupyter Notebook 可以单独运行每个cell, 并方便显示可视化结果, 请尽量使用jupyter notebook
2. Jupyter Notebook 不能并行, 请不要开并行(不过我觉得也用不到)
3. Jupyter Notebook的快捷键和用法可以自行百度, 可以在notebook里写Markdown(就像我现在正在做的那样)

### 使用小标题和注释
1. 每个cell(代码块)尽量只做一件事情, 并且在代码块的开始以注释的形式写明这个代码块在做什么(例如: #画出所有特征的分布图)
2. 灵活使用Markdown的标题, 比如一级大标题-数据预处理, 二级标题-特征描述, 二级标题-相关性分析 等等
3. 具体的实现细节不需要写注释, 按照你自己的习惯就好

### 文字分析
1. 一定要给出对结果的文字分析, 不要以注释的方式写, 而是应该在code下面新建一个Markdown cell
2. 如果有什么比较重要的结论, 可以直接在群里说, 也可以在Attention文件里记录. 大家注意一下信息的交流
